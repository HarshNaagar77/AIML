## Ploting data 

```python
data1 = {
    "Salary" : [ 20000 , 40000, 15000 , 25000 , 50000 , 60000 , 30000 , 28000 , 20000 , 35000]
}
df = pd.DataFrame(data1)
```

```python
plt.plot(df["Salary"] , color = "red" , marker = "o")
plt.grid()
```

## Histogram

```python
plt.hist(df["Salary"] , bins= 5 , color= "Blue" )
```

## Box Plot

```python
plt.boxplot(df["Salary"])
plt.grid()
```

## Pie chart

```python
df["Dept"] = ["HR" , "IT" , "FIN" , "HR" , "ENG"] * 2
df
```

```python
count = df.value_counts(df["Dept"])
count
```

```python
plt.pie(count , labels= count.index , autopct= "%1.1f" , explode = [0,0.1 , 0 ,0])
```

```python
count2 = df.value_counts(df["Salary"])
print(plt.pie(count2 , labels= count2.index , autopct= "%1.00f"))
```
