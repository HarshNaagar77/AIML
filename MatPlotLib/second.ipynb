```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Count plot     

```python
df = pd.read_csv("data/first.csv")
```

```python
df
```

```python
count = df.value_counts(df["Dept"])
plt.bar(count.index , count)
```

## Bivariate

```python

df["Age"] = np.random.randint(20, 50, size=len(df))
df
```

#### Bivariate refers to the plotting of bar with respect to two values like age and salary etc

## Scatter plot

```python
plt.scatter(df["Age"] , df["Salary"] , color = "Red")
```

## Line plot

```python
plt.plot( df["Age"] , df["Salary"] , marker = "*" , linestyle = ":" , linewidth = "2"  , color = "red")
```

```python
sortedAge = df.sort_values("Age")
```

```python
plt.plot( sortedAge["Age"] , df["Salary"] , marker = "*" , linestyle = ":" , linewidth = "2"  , color = "red")
```

## Bar chart

```python
plt.bar(df["Age"] , df["Salary"] , color = "red" , align= "center")
```

## Bivariate with one numerical and one categorical

```python
hr = df[df["Dept"] == "HR"]["Salary"]
print(hr)

it = df[df["Dept"] == "IT"]["Salary"]
it
```

```python
plt.boxplot( [hr , it])
```
