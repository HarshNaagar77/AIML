## Save and load

```python
arr1 = np.array([2,3,4,5,1,4])
arr2 = np.floor( np.random.random((3,3))*100 )

np.save('./arrays/array1.npy' , arr1)
np.save('./arrays/array2.npy' , arr2)
```

```python
loadedArr1 = np.load("./arrays/array1.npy")
print(loadedArr1)
```

```python
arr2d = np.random.random((3,3)) * 100
print(arr2d)
print("\n")

ans = np.where(arr2d > 50)
print(arr2d[ans])
print("\n")

ans2 = np.where( arr2d > 50 , "selected" , "rejected")
print(ans2)
```
