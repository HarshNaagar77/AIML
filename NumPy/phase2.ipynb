## NumPy Array Operations

```python
import numpy as np
```

```python
#slicing
arr = np.array([3, 4,2 ,6 ,4,4 ,7,4])
print(arr[2:5])
```

```python
#indexing in 2d array
arr2d = np.floor(np.random.random((3,3))* 100)
print(arr2d)

print(arr2d[1,2])    # 2nd index element of index 1 row
print(arr2d[1])      # entire 1st index row
print(arr2d[:,2])   # entire 2nd index column
```

## Sorting

```python
# 1d array sorting 
print("sorted array:", np.sort(arr))

# 2d array sorting
print("original 2d array:\n", arr2d)
print("sorted 2d array along rows:\n", np.sort(arr2d, axis=1))
print("sorted 2d array along columns:\n", np.sort(arr2d, axis=0))
```

## Filter

```python
# In numpy , arrays can also use mask i.e expressions for indexing

newArr = np.array([10, 15, 20, 25, 30, 35, 40])
mask = newArr > 25
print(newArr[mask])

# here we passed an expression as index in an array which lead to solving the expression and gave us output
```

## Where 

```python
# we can use "where" to implement a condition just like mask

res = np.where(newArr > 20)
print(newArr[res])
```

```python
# we can say that "where" works as if else clause
# here , we write it as 
# if newArr > 20 
# print newArr*2
# else newArr

res2 = np.where( newArr > 20 , newArr * 2 , newArr )
print(res2)
```
