## Advance operations with examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print(sales_data[0:4])  # prints the top 4 data
print("\n")

print(sales_data[:,1:])     # prints all rows and columns from index 1 onwards
```

```python
# print sum of array column wise
sum = np.sum(sales_data[: , 1:] , axis=0)
print(sum)
```

```python
sum_per_year = np.sum(sales_data[: , 1:3] , axis=0)
print(sum_per_year)
```

```python
# min value per restraunt
min = np.min(sales_data[: , 1:] , axis=1)
print(min)
```

```python
# max value per restraunt
max = np.max(sales_data[ : , 1:] , axis=1)
print(max)
```

```python
print( np.mean( sales_data[ : , 1:] , axis=0))
```

```python
print( np.cumsum( sales_data[ : , 1:] , axis=1))
```

## Plotting values on graph using matplotlib

```python
plt.Figure( figsize=(1,1))
print(sum)
plt.plot(sum)
plt.title(" Total sales of all restraunts")
plt.show()
```

```python
plt.Figure( figsize=(8,6))
print(min)
plt.plot(min)
plt.title(" minimum sales of every restraunts")
plt.show()
```

# Vector operations like add , multiply etc

### matrix multiplication

```python
vec1 = np.array([[1 ,2 ,3] , [1 , 2,3]])
print(vec1)
print("\n")
vec2 = np.array([ [1 ,1] , [2 ,2] , [3,3]])
print(vec2)
print("\n")
print(np.matmul(vec1 , vec2))
```

### Dot product

```python
print( np.dot(vec1 , vec2))
```

# Vectorization 

##### As we know that there are multiple methods in native python like upper(), lower() , split() etc. But we can not use them directly in numpy , so to use them we must first vectorize them. After that we can use them in numpy normally.

```python
vec3 = np.array( ["Chinese" , "Italian" , "Indian" , "American"])
vectorize_method1 = np.vectorize(str.upper)

print( vectorize_method1(vec3))
```
