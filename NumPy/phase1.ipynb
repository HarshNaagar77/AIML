## Numpy arrays and basics

```python
import numpy as np
```

## Creating arrays from list

```python
arr1D = np.array([1, 2, 3, 4, 5])
print(arr1D)
```

```python
arr2D = np.array([[1,3,5] , [2,4,6]])
print(arr2D)
```

```python
list1 = [10, 20, 30, 40, 50]
print(list1)

print(list1 * 2)  # This will repeat the list elements i.e multiplication is list is not element wise
```

```python
print(arr1D * 2)  # This will multiply each element by 2 i.e element wise multiplication
```

### List vs Array in np which is faster

```python
import time

time1 = time.time()
list2 = [ i*2 for i in range(1, 1000000)]
print("Time taken by list comprehension: ", time.time() - time1)
```

```python
time1 = time.time()
arr3= np.arange(1, 1000000) * 2   # Arange creates an array with values from start to end-1
print("Time taken by numpy array operation: ", time.time() - time1)
```

### Creating matrices of any number

```python
zero = np.zeros((3,4))  # 3 rows and 4 columns
print("Matrix of zeros:\n", zero)
```

```python
one = np.ones((2,5))
print("Matrix of ones:\n", one)
```

```python
# Matrix of any number
mat = np.full((3,3) , 8)
print("Matrix of 8s:\n", mat)
```

```python
#Matrix of random number 
random = np.random.random((3,3))
print("Matrix of random numbers:\n", random * 100)
```

### Array Reshaping

```python
arr4 = np.arange(12)
print("Original array:\n", arr4)

reshaped = arr4.reshape((3,4))
print("Reshaped array (3x4):\n", reshaped)

flatend = reshaped.flatten()                # Flatten returns a flattened array as a copy   
print("Flattened array:\n", flatend)

raveled = reshaped.ravel()                  # Ravel returns a flattened array but as a original
print("Raveled array:\n", raveled)

transposed = reshaped.T
print("Transposed array:\n", transposed)
```
