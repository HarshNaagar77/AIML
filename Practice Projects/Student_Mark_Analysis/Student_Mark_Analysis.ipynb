## Students Marks Analysis

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/marksheet.csv")
data
```

```python
min = np.min(data["Age"], axis=0 )
print(min)
```

```python
max = np.max(data["Age"], axis=0 )
print(max)
```

```python
data.iloc[: , 6:].sum(axis=1)
```

```python
data["Total"] = data[ ["Science" ,"English" , "History" , "Maths"]].sum(axis=1)
data
```

```python
data[ ["Science" ,"English" , "History" , "Maths"]].mean( axis=1)
```

```python
# adding average column

data["Average"] = data[ ["Science" ,"English" , "History" , "Maths"]].mean(axis=1)
data
```

```python
# Sorting by average

avg = data["Average"].sort_values(axis=0).index
n = data.loc[ avg , ["Name" , "Average"]]
n
```
